```python
data2=data[(data["Event distance/length"].isin(["50km","50mi"])) & (data["Year of event"]==2020) & (
    data["Event name"].str.split("(").str.get(1).str.split(")").str.get(0)=="USA")]
```

Removing USA from event name for convenience

```python
data2['Event name'].str.split("(").str.get(0)
```

Changing Column values to ensure no (USA) is present

```python
data2['Event name']=data2['Event name'].str.split("(").str.get(0)
```

```python
data2.head()
```

Cleaning Up athelete age

```python
data2["athlete_age"]=2020-data2["Athlete year of birth"]
```

```python
data2.head()
```

Removing "h" from athlete performance

```python
data2["Athlete performance"]=data2["Athlete performance"].str.split(" ").str.get(0)
```

Dropping columns which arent necessary (for now)

```python
data2=data2.drop(["Athlete club",'Athlete country','Athlete year of birth','Athlete age category'],axis=1)
```

Rename Columns

```python
"""**Year of event                  int64

Event dates                   object

Event name                    object

Event distance/length         object

Event number of finishers      int64

Athlete performance           object

Athlete gender                object

Athlete average speed        float64

Athlete ID                     int64

athlete_age                    int64
**
"""
```

```python
data2=data2.rename(columns={"Year of event":"Year","Event dates":"Race_day","Event name":"Race_name",
                            "Event distance/length":"Race_length","Event number of finishers":"Race_Number_of_finishers",
                            "Athlete performance":"Athlete_Performance","Athlete gender":"Athlete_Gender",
                            "Athlete average speed":"Athlete_Avg_Speed","Athlete ID":"Athlete_ID"
    ,"athlete_age":"Athlete_Age"
})
```

```python
data2.head()
```

Reorder columns

```python
data3=data2[["Race_day","Race_name"	,"Race_length","Race_Number_of_finishers","Athletle_ID","Athlete_Gender","Athlete_Age","Athlete_Performance","Athlete_Avg_Speed"]]
```

Check for difference in speed between Genders for the 50MI,50KM

```python
data3.groupby(["Race_length","Athlete_Gender"])["Athlete_Avg_Speed"].mean()
```

Means are closer to each other implying almost similar performance in 50km and a bit lesser performance from Females in the 50mi category

Q)Which age groups perform the best in 50mi race?

```python
data3.query('Race_length=="50mi"').groupby("Athlete_Age")["Athlete_Avg_Speed"].agg(['mean','count']).sort_values('mean',ascending=False).query('count>19').head(15)
```

Q)Which age groups perform the worst in 50mi race?

```python
data3.query('Race_length=="50mi"').groupby("Athlete_Age")["Athlete_Avg_Speed"].agg(['mean','count']).sort_values('mean').query('count>10').head(20)
```

Seperate acc to seasons

```python
data3["Race_Month"]=data3["Race_day"].str.split(".").str.get(1).astype(int)
```

```python
data3["Race_Season"]=data3['Race_Month'].apply(lambda x:'Winter' if x>11 else 'Fall' if x>8 else 'Summer' if x>5 else 'Spring')
```

```python
data3.head()
```

```python
data3.groupby('Race_Season')["Athlete_Avg_Speed"].agg(["mean",'count']).sort_values("mean",ascending=False)
```

Now only for 50mi

```python
data3.query('Race_length=="50mi"').groupby('Race_Season')["Athlete_Avg_Speed"].agg(["mean",'count']).sort_values("mean",ascending=False)
```
